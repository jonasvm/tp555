```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Vetores x1 e x2 com valores uniformemente espaçados entre -10 e 11 com passos de 0.25 unidades
x1 = np.arange(-10, 11, 0.25)
x2 = np.arange(-10, 11, 0.25)

# Criando uma malha de coordenadas (grid) para calcular os valores de y
X1, X2 = np.meshgrid(x1, x2)

# Calculando y = x1^2 + x2^2
Y = X1**2 + X2**2

# Figura e gráfico 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plotando a superfície 3D
ax.plot_surface(X1, X2, Y, cmap='viridis')

# Definindo os rótulos dos eixos
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')

# Gráfico
plt.show()
```
