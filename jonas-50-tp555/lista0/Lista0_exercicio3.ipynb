```python
import numpy as np
import matplotlib.pyplot as plt

M = 1000

# Vetor coluna com M elementos retirados de uma distribuição aleatória uniforme com valores no intervalo em [0, 1)
x = np.random.uniform(0, 1, M)

# Vetor coluna com M elementos retirados de uma distribuição aleatória Gaussiana normal, i.e., com média 0 e variância unitária.
w = np.random.normal(0, 1, M)

# Vetor coluna, y, com M = 1000 elementos, onde y é dado pela equação y = 1.2 + 2.3*x + 10*w
y = 1.2 + 2.3 * x + 10 * w

# Gráfico
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='b', marker='o', label='Pontos (x, y)')
plt.title('Gráfico 2D', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
