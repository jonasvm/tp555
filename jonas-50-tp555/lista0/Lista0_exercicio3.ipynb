```python
import numpy as np
import matplotlib.pyplot as plt

M = 1000

# Vetor coluna com M elementos retirados de uma distribuio aleatria uniforme com valores no intervalo em [0, 1)
x = np.random.uniform(0, 1, M)

# Vetor coluna com M elementos retirados de uma distribuio aleatria Gaussiana normal, i.e., com mdia 0 e varincia unitria.
w = np.random.normal(0, 1, M)

# Vetor coluna, y, com M = 1000 elementos, onde y  dado pela equao y = 1.2 + 2.3*x + 10*w
y = 1.2 + 2.3 * x + 10 * w

# Grfico
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='b', marker='o', label='Pontos (x, y)')
plt.title('Grfico 2D', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
