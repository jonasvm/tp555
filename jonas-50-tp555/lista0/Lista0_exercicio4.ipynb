```python
import numpy as np
import matplotlib.pyplot as plt

M = 10000

# Vetores x e y com distribuição uniforme
x = np.random.uniform(size=M)
y = np.random.uniform(size=M)

# Calculando o vetor z
z = x + y

# Plotando o histograma normalizado de z
plt.hist(z, bins=50, density=True, alpha=0.7, color='blue')
plt.title('Histograma')
plt.xlabel('Z')
plt.ylabel('Densidade de Probabilidade')
plt.grid(True)
plt.show()
```
