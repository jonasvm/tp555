```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB, BernoulliNB

# Conjunto de treinamento com frequência das palavras no vocabulário V = ["good", "poor", "boring", "great"]
x_train = np.array([
    [3, 0, 0, 3],  # Positive (+)
    [0, 1, 1, 2],  # Positive (+)
    [1, 0, 1, 2],  # Positive (+)
    [1, 3, 2, 0],  # Negative (-)
    [1, 5, 0, 2],  # Negative (-)
    [0, 2, 2, 0]   # Negative (-)
])
y_train = np.array(['Positive', 'Positive', 'Positive', 'Negative', 'Negative', 'Negative'])

# Críticas de teste representadas como frequência das palavras ["good", "poor", "boring", "great"]
x_test = np.array([
    [2, 0, 1, 2],  # Critique 1
    [3, 1, 0, 1],  # Critique 2
    [1, 1, 0, 1]   # Critique 3
])

# 1. Treinamento dos classificadores Naive Bayes
# MultinomialNB
multinomial_model = MultinomialNB()
multinomial_model.fit(x_train, y_train)

# BernoulliNB (usa matriz binária)
vectorizer = CountVectorizer(binary=True)
x_train_binary = (x_train > 0).astype(int)  # Converte para binário (presença/ausência de palavras)
x_test_binary = (x_test > 0).astype(int)
bernoulli_model = BernoulliNB()
bernoulli_model.fit(x_train_binary, y_train)

# 2. Previsão das classes para as críticas de teste
y_pred_multinomial = multinomial_model.predict(x_test)
y_pred_bernoulli = bernoulli_model.predict(x_test_binary)

# 3. Exibir os resultados das previsões
print("Classe prevista pelo MultinomialNB para cada crítica de teste:", y_pred_multinomial)
print("Classe prevista pelo BernoulliNB para cada crítica de teste:", y_pred_bernoulli)

# 4. Exibir as probabilidades das classes para cada crítica de teste
print("\nProbabilidades de classe pelo MultinomialNB:")
print(multinomial_model.predict_proba(x_test))

print("\nProbabilidades de classe pelo BernoulliNB:")
print(bernoulli_model.predict_proba(x_test_binary))
```
